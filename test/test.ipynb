```python
# 初始化一个长度为1400的字节数组
data = bytearray(1400)

# 设置前两个字节为'H'和'L'的ASCII值
data[0] = ord('H')
data[1] = ord('L')

# 设置第三个字节为16进制的6
data[2] = 0x06

# 从第10个字节开始，每100个字节数据一致
# 这里我们选择一个固定的100字节数据块
# 你可以根据需要修改这个数据块的内容
pattern = bytes([0x01] * 100)  # 这里用0x01填充，你可以换成其他值

# 计算需要填充的完整100字节块的数量
num_full_blocks = (1400 - 10) // 100

# 填充完整100字节块
for i in range(num_full_blocks):
    start = 10 + i * 100
    end = start + 100
    data[start:end] = pattern

# 处理剩余的字节
remaining_bytes = (1400 - 10) % 100
if remaining_bytes > 0:
    start = 10 + num_full_blocks * 100
    end = start + remaining_bytes
    data[start:end] = pattern[:remaining_bytes]

# 打印前20个字节以验证
print("前20个字节：", data)

# 打印第10个字节开始的前10个字节以验证
print("第10个字节开始的前10个字节：", data[10:20])
```
